```python
logmodel.fit(X_train_minmax,Y_train)
```

```python
predictions = logmodel.predict(X_test_minmax)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,predictions))
```

# Kfold


```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn import tree
X = df3.drop('salary',axis=1)
Y = df3['salary']
```

```python
kf3 = KFold(n_splits=3,shuffle=True,random_state=42)

cnt=1

for train_index,test_index in kf3.split(X,Y):
    print(f'Fold:{cnt}, Train set: {len(train_index)}, Test set:{len(test_index)}')
    cnt += 1
        
```

```python
score  = cross_val_score(LogisticRegression(random_state=42),X,Y,cv=kf3,scoring='accuracy')
print(f"Scores for each fold are : {score}")
```

```python
print(f"Average Score:{score.mean()}")
```

```python
dt = tree.DecisionTreeClassifier(random_state=42)
score  = cross_val_score(dt,X,Y,cv=kf3,scoring='accuracy')
print(f"Scores for each fold are : {score}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
pipe = Pipeline([('classifier' , RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=None, max_features=6, max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=100, n_jobs=None,
            oob_score=False, random_state=None, verbose=0,
            warm_start=False))])
# pipe = Pipeline([('classifier', RandomForestClassifier())])

# Create param grid.

param_grid = [
    {'classifier' : [LogisticRegression()],
     'classifier__penalty' : ['l1', 'l2'],
    'classifier__C' : np.logspace(-4, 4, 20),
    'classifier__solver' : ['liblinear']},
    {'classifier' : [RandomForestClassifier()],
    'classifier__n_estimators' : list(range(10,101,10)),
    'classifier__max_features' : list(range(6,32,5))}
]

# Create grid search object

clf = GridSearchCV(pipe, param_grid = param_grid, cv = 5, verbose=True, n_jobs=-1)

# Fit on data

best_clf = clf.fit(X_train_minmax, Y_train)
```
