```python
import numpy as np 
import pandas as  pd 

cols = ["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","salary"]
df = pd.read_csv("adult.data", names=cols,skipinitialspace=True)
```

```python
df
```

```python
df = df.replace(0,np.nan)
    
```

```python
df.isna().sum()/df.shape[0]*100
```

```python
df = df.drop(columns = ['capital-gain','capital-loss'])
df
```

```python
df['fnlwgt'].nunique()
```

```python
df = df.drop('fnlwgt',axis=1)
```

```python
df
```

```python
df['salary'].value_counts()
```

```python
df1=df.copy()
df1
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
df1['relationship'] = class_le.fit_transform(df1['relationship'].values)
```

```python
df1
```

```python
salary_mapping = {'<=50K':0,'>50K':1}
df1['salary'] = df1['salary'].map(salary_mapping)
```

```python
df1[['relationship','salary']].corr()
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
df1['education'] = class_le.fit_transform(df1['education'].values)
```

```python
df1
```

```python
race_mapping = {'White':0,'Black':1,'Asian-Pac-Islander':2,'Amer-Indian-Eskimo':4,'Other':5}
df1['race'] = df1['race'].map(race_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
df1['occupation'] = class_le.fit_transform(df1['occupation'].values)
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
df1['native-country'] = class_le.fit_transform(df1['native-country'].values)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
df1['workclass'] = class_le.fit_transform(df1['workclass'].values)
```

```python
df1
```

```python
sex_mapping = {'Male':0,'Female':1}
df1['sex'] = df1['sex'].map(sex_mapping)
```

```python
df1['edu_occ'] = df1['education-num']+df1['occupation']
```

```python
df1[['edu_occ','salary']].corr()
```

```python
df1.drop('marital-status',axis=1,inplace=True)
```

```python
df1
```

```python
df1['age_hr'] = df1['age']+df1['hours-per-week']
```

```python
df1[['age_hr','salary']].corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
X = df1.drop('salary',axis=1)
Y = df1['salary']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

# Random Forest

```python
rfc =RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=20,max_features='auto',min_samples_leaf=10,min_samples_split=50)
```

```python
rfc.fit(X_train,Y_train)
```

```python
rfc_predict = rfc.predict(X_test)
```

```python
print(classification_report(Y_test,rfc_predict))
```

# Ada Boost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(n_estimators=500,learning_rate=.5)
```

```python
ada.fit(X_train,Y_train)
```

```python
ada_predict = ada.predict(X_test)
```

```python
print(classification_report(Y_test,ada_predict))
```

# CatBoost 

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier(iterations=1000, 
                           task_type="GPU",
                           devices='0:1',
                          learning_rate=.05,)
```

```python
model.fit(X_train,
          Y_train,
          verbose=False)
```

```python
model_predict = model.predict(X_test)
```

```python
print(classification_report(Y_test,model_predict))
```
