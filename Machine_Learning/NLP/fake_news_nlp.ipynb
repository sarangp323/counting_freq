```python
df=df.dropna()
```

```python
messages = df.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
df
```

```python
from nltk.corpus import stopwords

import string
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
count=0
def text_process(news):
    global count
    count+=1
    print(count,end="\r")
    nopunc = [word for word in news if word not in string.punctuation]
    nopunc = "".join(nopunc)
    return " ".join(ps.stem(word) for word in nopunc.split() if word.lower() not in stopwords.words('english'))
```

```python
messages['clean_text']= messages['text'].apply(text_process)
```

```python
messages
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer(ngram_range=[1,1],
                             tokenizer=lambda x: x.split(),
                             binary=True)
```

```python
X = tf.fit_transform(messages['clean_text'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,messages['label'],test_size=.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive_model = MultinomialNB(alpha=.01).fit(X_train,y_train)
```

```python
prediction = naive_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```
