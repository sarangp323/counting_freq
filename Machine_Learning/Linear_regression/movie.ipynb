```python
def check_weekday(date):
      
    # computing the parameter date
    # with len function
    res=len(pd.bdate_range(date,date))
      
    if res == 0 :
        return 'weekend'
    else:
        return 'weekday' 
```

```python
movie['weeknd'] = movie['release_date'].apply(check_weekday)
```

```python
movie
```

```python
def friday(date):
    temp = pd.Timestamp(date)
    if temp.day_name()=='Friday':
        return 1
    else:
        return 0
```

```python
movie['friday'] = movie['release_date'].apply(friday)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
rg = Ridge(random_state=42)
```

```python
rg.fit(X_train_minmax,y_train)
```

```python
rg_pred = rg.predict(X_test_minmax)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, rg_pred))
print('MSE:', metrics.mean_squared_error(y_test, rg_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rg_pred)))
```

```python
grid = dict()
grid['alpha'] = np.arange(0, 1, 0.01)
grid['tol']=[.001,.00001,.002,.0001,.0002]
grid['fit_intercept']=[True,False]
grid['solver']=['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
```

```python
from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.model_selection import RepeatedKFold
```

```python
rg_tune=Ridge(random_state=42)
```

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=42)
```

```python
rg_model = GridSearchCV(rg_tune,param_grid=grid ,cv=cv, scoring='neg_mean_squared_error',n_jobs=-1)
```

```python
rg_result = rg_model.fit(X_train_minmax, y_train)
```

```python
rg_result.best_score_
```
